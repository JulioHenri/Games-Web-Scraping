```
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
from urllib.request import urlopen
```

```
#Declarando variáveis
games = []

#Obtendo o HTML e o total de páginas
url = 'https://www.gamespot.com/games/reviews/'
response = urlopen(url)
html = response.read()
soup = BeautifulSoup(html, 'html.parser')
pages = int(soup.find('li',{'class':'paginate__item paginate__ellipse'}).findNextSibling().get_text())

for i in range(pages):
  #Obtendo o HTML
  response = urlopen('https://www.gamespot.com/games/reviews/?page=' + str(i+1))
  html = response.read()
  soup = BeautifulSoup(html, 'html.parser')

  #Obtendo TAGS de interesse
  cards = soup.find('section', {'class':'editorial river '}).findAll('div',{'class':'horizontal-card-item '})

  #Iterando sob todos os cards da página
  for card in cards:
    game = {}
    #Rating
    v = card.find('div',{'class':'review-ring-score__score text-bold '})
    if v:
      game['rating'] = v.get_text()
    else:
      game['rating'] = 'N/A'
    #Game
    game['name'] = card.find('h4', {'class':'horizontal-card-item__title'}).get_text()
    #Release date
    game['release_date'] = card.find('div', {'class':'card-metadata'}).find('time').get_text()
        
    numeros = []
    itens = card.find('div', {'class':'card-metadata'}).findAll('span')
    for item in itens:
      numeros.append(item.get_text())

    #Comment
    game['comment'] = numeros[0]
    #Like
    game['like'] = numeros[1]
    #Platform
    v = card.find('span',{'class':'horizontal-card-item__label text-bold text-small'})
    if v:
      game['platform'] = v.get_text()
    else:
      game['platform'] = 'N/A'
    #Adicionando resultado à lista games
    games.append(game)

df = pd.DataFrame(games)
df.head()
```
