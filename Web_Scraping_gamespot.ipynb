# Iterando para todos os cards da primeira página

```
url = 'https://www.gamespot.com/games/reviews/'

response = urlopen(url)
html = response.read()
html = html.decode('utf-8')
html = tratar_html(html)
soup = BeautifulSoup(html, 'html.parser')


#Declarando variáveis
games = []
#Obtendo TAGS de interesse
cards = soup.find('div', {'id':'js-sort-filter-results'}).findAll('div',{'class':'horizontal-card-item'})

#Iterando sob todos os cards da página
for card in cards:
  game = {}
  #Rating
  game['rating'] = card.find('div',{'class':'review-ring-score__score text-bold '}).getText()
  #Platform
  game['platform'] = card.find('span',{'class':'horizontal-card-item__label text-bold text-small'}).getText()
  #Game
  game['name'] = card.find('h4', {'class':'horizontal-card-item__title'}).get_text()
  #Release date
  game['release_date'] = card.find('div', {'class':'card-metadata'}).find('time').get_text()

  numeros = []
  itens = card.find('div', {'class':'card-metadata'}).findAll('span')
  for item in itens:
    numeros.append(item.get_text())

  #Comment
  game['comment'] = numeros[0]
  #Like
  game['like'] = numeros[1]

  #Adicionando resultado à lista games
  games.append(game)
#Criando um df com o resultado
df = pd.DataFrame(games)
df
```

# Iterando para todas as páginas do site

Obtendo o número total de páginas.

```
int(soup.find('a',{'href':'/games/reviews/?page=717'}).get_text())
```

```
#número total de páginas
int(soup.find('li',{'class':'paginate__item paginate__ellipse'}).findNextSibling().get_text())
```

```
pages = int(soup.find('a',{'href':'/games/reviews/?page=717'}).get_text())
for i in range(pages):
  response = urlopen('https://www.gamespot.com/games/reviews/?page=' + str(i+1))
  html = response.read()
  soup = BeautifulSoup(html, 'html.parser')
  #Declarando variáveis
  games = []
  #Obtendo TAGS de interesse
  cards = soup.find('div', {'id':'js-sort-filter-results'}).findAll('div',{'class':'horizontal-card-item'})

  #Iterando sob todos os cards da página
  for card in cards:
    game = {}
    #Game
    game['name'] = card.find('h4', {'class':'horizontal-card-item__title'}).get_text()
    #Rating
    v = card.find('div',{'class':'review-ring-score__score text-bold '})
    if v:
      game['rating'] = v.get_text()
    else:
      game['rating'] = 'N/A'    #Release date
    game['release_date'] = card.find('div', {'class':'card-metadata'}).find('time').get_text()
    #Platform
    v = card.find('span',{'class':'horizontal-card-item__label text-bold text-small'})
    if v:
      game['platform'] = v.get_text()
    else:
      game['platform'] = 'N/A'    

    numeros = []
    itens = card.find('div', {'class':'card-metadata'}).findAll('span')
    for item in itens:
      numeros.append(item.get_text())

    #Comment
    game['comment'] = numeros[0]
    #Like
    game['like'] = numeros[1]

    #Adicionando resultado à lista games
    games.append(game)
  #Criando um df com o resultado
df = pd.DataFrame(games)
df
```
